```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
device
```

```python
model_name = "openai-community/gpt2-large"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
```

```python
input_text = " It was pretty good"
max_length = 128
input_ids = tokenizer(input_text, return_tensors="pt")
```

```python
input_ids
```

```python
input_ids = input_ids["input_ids"].to(device)
```

```python
output = model.generate(input_ids, max_length=max_length, num_beams=5, do_sample=False)
```

```python
output
```

```python
print(tokenizer.decode(output[0]))
```

```python
output = model.generate(input_ids, max_length=max_length, num_beams=5, do_sample=False, no_repeat_ngram_size=2)
tokenizer.decode(output[0])
```

### Nucleus Search


```python
input_text = " It was pretty good  "
max_length = 128
input_ids = tokenizer(input_text, return_tensors="pt")
input_ids = input_ids["input_ids"].to(device)

output = model.generate(input_ids, max_length=max_length, num_beams=5, do_sample=False, top_p=0.9)
print(tokenizer.decode(output[0]))
```
